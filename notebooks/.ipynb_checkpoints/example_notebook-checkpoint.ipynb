```python
#Missing Values Function
from sklearn.ensemble import RandomForestRegressor

# Function for comparing different approaches
def score_dataset(X_train_full, X_test_full, y_train_full):
    model = RandomForestRegressor(n_estimators=100, random_state=0)
    model.fit(X_train_full, y_train_full)
    
    preds = model.predict(X_test_full)
    model_assessment= model.score(X_train_full, y_train_full)
    acc_random_forest = round(model_assessment*100, 2)
    print(round(acc_random_forest,2,), "%")
```

```python
# Get names of columns with missing values
cols_with_missing = [col for col in X_train_full.columns
                     if X_train_full[col].isnull().any()]

# Drop columns in training and validation data
reduced_X_train = X_train_full.drop(cols_with_missing, axis=1)
reduced_X_test = X_test_full.drop(cols_with_missing, axis=1)

print("RandomForestRegressor Score from Approach 1 (Drop columns with missing values):")
print(score_dataset(reduced_X_train, reduced_X_test, y_train_full))
```
